```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white')
```

```python
df = pd.read_csv("Datasets/data_delimited.csv",index_col=0)
```

Las variables categóricas son:
- track.explicit
- key
- mode
- time_signature

```python
counts = df['track.explicit'].value_counts();
as_percentages = counts * 100 / counts.sum();

ax = sns.barplot(as_percentages, as_percentages.index, orient='h');
ax.set_title('Explicitud');
ax.set(xlabel='Porcentaje total');
```

```python
counts = df['key'].value_counts();
as_percentages = counts * 100 / counts.sum();

ax = sns.barplot(as_percentages, as_percentages.index, orient='h');
ax.set_title('Tonalidad');
ax.set(xlabel='Porcentaje total');
```

```python
counts = df['mode'].value_counts();
as_percentages = counts * 100 / counts.sum();

ax = sns.barplot(as_percentages, as_percentages.index, orient='h');
ax.set_title('Modalidad');
ax.set(xlabel='Porcentaje total');
```

```python
counts = df['time_signature'].value_counts();
as_percentages = counts * 100 / counts.sum();

ax = sns.barplot(as_percentages, as_percentages.index, orient='h');
ax.set_title('Signatura');
ax.set(xlabel='Porcentaje total');
```

### Análisis multivariable

```python
crosstab = pd.crosstab(df['key'], df['track.explicit'])

fig, axes = plt.subplots(3, 4, figsize=(10, 8), sharex=True, sharey=True)

sns.barplot(crosstab.columns, crosstab.loc[1], ax=axes[0, 0])
sns.barplot(crosstab.columns, crosstab.loc[2], ax=axes[0, 1])
sns.barplot(crosstab.columns, crosstab.loc[3], ax=axes[0, 2])
sns.barplot(crosstab.columns, crosstab.loc[4], ax=axes[0, 3])
sns.barplot(crosstab.columns, crosstab.loc[5], ax=axes[1, 0])
sns.barplot(crosstab.columns, crosstab.loc[6], ax=axes[1, 1])
sns.barplot(crosstab.columns, crosstab.loc[7], ax=axes[1, 2])
sns.barplot(crosstab.columns, crosstab.loc[8], ax=axes[1, 3])
sns.barplot(crosstab.columns, crosstab.loc[9], ax=axes[2, 0])
sns.barplot(crosstab.columns, crosstab.loc[10], ax=axes[2, 1])
sns.barplot(crosstab.columns, crosstab.loc[11], ax=axes[2, 2])



axes[0, 0].set(xlabel='', ylabel='', title='Modo: 1')
axes[0, 1].set(xlabel='', ylabel='', title='Modo: 2')
axes[0, 2].set(xlabel='', ylabel='', title='Modo: 3')
axes[0, 3].set(xlabel='', ylabel='', title='Modo: 4')
axes[1, 0].set(xlabel='', ylabel='', title='Modo: 5')
axes[1, 1].set(xlabel='', ylabel='', title='Modo: 6')
axes[1, 2].set(xlabel='', ylabel='', title='Modo: 7')
axes[1, 3].set(xlabel='', ylabel='', title='Modo: 8')
axes[2, 0].set(xlabel='', ylabel='', title='Modo: 9')
axes[2, 1].set(xlabel='', ylabel='', title='Modo: 10')
axes[2, 2].set(xlabel='', ylabel='', title='Modo: 11')

fig.suptitle('Explicitud dependiendo del tono de las canciones', fontsize=15);
```
